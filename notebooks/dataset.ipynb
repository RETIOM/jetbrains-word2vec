```python
import re
from collections import Counter

import numpy as np
```

```python
with open("../data/wikitext-103/wiki.train.tokens") as f:
    data = f.read()

text = data.lower()
text = re.sub(r"[^a-z0-9]+", " ", text)
text = text.split()

counts = Counter(text)
vocab = {w: c for w, c in counts.items() if c >= 5}

word2idx = {w: i for i, w in enumerate(vocab.keys())}
idx2word = {i: w for w, i in word2idx.items()}
```

```python
test = data.lower()
test = re.sub(r"^ =.*= $", "", test, flags=re.MULTILINE)  # Remove headings
test = test.split("\n\n")  # Split into paragraphs

for par in test:
    txt = re.sub(r"[^a-z0-9]+", " ", par.strip())
    if not txt:
        continue
```

```python
re.sub(r"[^a-z0-9]+", " ", test.split("\n\n")[4])
```

```python
word_ids = np.array([word2idx.get(word, -1) for word in text])
word_ids = word_ids[word_ids != -1]
```

```python
test_words = word_ids[:20]

window_size = 2
for idx, target in enumerate(test_words):
    start = max(idx - window_size, 0)
    end = min(idx + window_size + 1, len(test_words) - 1)
    context = np.concatenate((test_words[start:idx], test_words[idx + 1 : end]))
    print((target, context))
```
